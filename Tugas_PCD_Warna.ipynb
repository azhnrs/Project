```python
import matplotlib, cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

%matplotlib inline
```

```python
img1 = cv2.imread('beach.jpg')
im1 = mpimg.imread('beach.jpg')
plt.subplot(121)
plt.imshow(im1)
plt.title('Gambar 1')

img2 = cv2.imread('plane.jpg')
im2 = mpimg.imread('plane.jpg')
plt.subplot(122)
plt.imshow(im2)
plt.title('Gambar 2')

plt.show()
```

```python
img1 = cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
plt.subplot(121)
plt.imshow(img1)
plt.title('Gambar 1')

img2 = cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
plt.subplot(122)
plt.imshow(img2)
plt.title('Gambar 2')

plt.show()
```

```python
foto_rgb = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)

    # Pisahkan saluran warna
red_channel = foto_rgb.copy()
red_channel[:, :, 1:] = 0  # Set hijau dan biru menjadi 0

green_channel = foto_rgb.copy()
green_channel[:, :, [0, 2]] = 0  # Set merah dan biru menjadi 0

blue_channel = foto_rgb.copy()
blue_channel[:, :, :2] = 0  # Set merah dan hijau menjadi 0

    # Tampilkan hasil
plt.figure(figsize=(12, 4))

plt.subplot(131)
plt.imshow(red_channel)
plt.title('Citra Red Channel')
plt.axis('off')

plt.subplot(132)
plt.imshow(green_channel)
plt.title('Citra Green Channel')
plt.axis('off')

plt.subplot(133)
plt.imshow(blue_channel)
plt.title('Citra Blue Channel')
plt.axis('off')
```

```python
if img1.shape == img2.shape:
    # Metode 1: Menggunakan cv2.add()
    hasil_add = cv2.add(img1, img2)

    # Metode 2: Menggunakan operator +
    hasil_operator = img1 + img2

    # Tampilkan gambar asli dan hasil penjumlahan
    cv2.imshow('Gambar 1', img1)
    cv2.imshow('Gambar 2', img2)
    cv2.imshow('Hasil Add', hasil_add)
    cv2.imshow('Hasil Operator', hasil_operator)
    cv2.waitKey(0)

else:
    print("Ukuran gambar tidak sama.")
```

```python


# Resize gambar agar memiliki ukuran yang sama
img1_resized = cv2.resize(img1, (img2.shape[1], img2.shape[0]))

# Jumlahkan kedua gambar
hasil_add = cv2.add(img1_resized, img2)

# Tampilkan gambar asli, gambar yang sudah diresize, dan hasil penjumlahan
plt.figure(figsize=(10, 5))

plt.subplot(131)
plt.imshow(cv2.cvtColor(img1, cv2.COLOR_BGR2RGB))
plt.title('Gambar 1')

plt.subplot(132)
plt.imshow(cv2.cvtColor(img2, cv2.COLOR_BGR2RGB))
plt.title('Gambar 2')

plt.subplot(133)
plt.imshow(cv2.cvtColor(hasil_add, cv2.COLOR_BGR2RGB))
plt.title('Hasil Add')

plt.show()
```
