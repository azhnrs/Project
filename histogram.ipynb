```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def plot_histogram(image):
    color = ('b','g','r')

    plt.figure(figsize=(15, 5))
    for i,col in enumerate(color):
        if len(image.shape) == 3:
            histr = cv2.calcHist([image],[i],None,[256],[0,256])
        else:
            histr = cv2.calcHist([image],[0],None,[256],[0,256])
        plt.subplot(1, 3, i+1)
        plt.bar(np.arange(256), histr.flatten(), color = col, alpha=0.5)
        plt.title(col.upper() + ' Channel')
        plt.xlabel('Intensity')
        plt.ylabel('Frequency')

    plt.show()

imageName = "ci.jpg"
image = cv2.imread(imageName)

plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title('Original Image')
plt.show()

plot_histogram(image)

# Histogram Equalization
image_eq = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
image_eq = cv2.equalizeHist(image_eq)

plt.imshow(image_eq, cmap='gray')
plt.axis('off')
plt.title('Equalized Image')
plt.show()

plot_histogram(image_eq)
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def plot_histogram(image, title):
    plt.figure(figsize=(8, 5))
    plt.hist(image.flatten(), bins=256, range=[0,256], color='k', alpha=0.5)
    plt.title(title)
    plt.xlabel('Intensity')
    plt.ylabel('Frequency')
    plt.show()

imageName = "ci.jpg"
image = cv2.imread(imageName)

plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title('Original Image')
plt.show()

# Convert to grayscale and equalize histogram
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
image_eq = cv2.equalizeHist(image_gray)

plt.imshow(image_eq, cmap='gray')
plt.axis('off')
plt.title('Equalized Image')
plt.show()

# Plot histogram after equalization
plot_histogram(image_eq, 'Histogram after Equalization')
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def plot_histogram(image, title):
    plt.figure(figsize=(8, 5))
    plt.hist(image.flatten(), bins=256, range=[0,256], color='k', alpha=0.5)
    plt.title(title)
    plt.xlabel('Intensity')
    plt.ylabel('Frequency')
    plt.show()

imageName = "ci.jpg"
image = cv2.imread(imageName)

plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title('Original Image')
plt.show()

# Convert to grayscale
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Apply CLAHE
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
image_clahe = clahe.apply(image_gray)

plt.imshow(image_clahe, cmap='gray')
plt.axis('off')
plt.title('CLAHE Equalized Image')
plt.show()

# Plot histogram after CLAHE equalization
plot_histogram(image_clahe, 'Histogram after CLAHE Equalization')
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def plot_histogram(image):
    gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
    hist = cv2.calcHist([gray], [0], None, [256], [0,256])

    plt.figure(figsize=(10, 5))
    plt.subplot(1, 2, 1)
    plt.imshow(gray, cmap='gray')
    plt.axis('off')
    plt.title('Grayscale Image')

    plt.subplot(1, 2, 2)
    plt.bar(np.arange(256), hist.flatten(), color='black', alpha=0.5)
    plt.title('Grayscale Histogram')
    plt.xlabel('Intensity')
    plt.ylabel('Frequency')

    plt.show()

imageName = "ci.jpg"
image = cv2.cvtColor(cv2.imread(imageName), cv2.COLOR_BGR2RGB)

plot_histogram(image)
```
