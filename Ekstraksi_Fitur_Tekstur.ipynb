# 2. Berikanlah contoh penggunaan ekstraksi fitur tekstur dari data input!

```python
from skimage.feature import greycomatrix, greycoprops
import numpy as np
from matplotlib import pyplot as plt

# Masukkan Gambar
image1 = plt.imread('1.jpg')
image2 = plt.imread('2.jpg')

image1_gray = np.uint8(0.2989 * image1[:,:,0] + 0.5870 * image1[:,:,1] + 0.1140 * image1[:,:,2])
image2_gray = np.uint8(0.2989 * image2[:,:,0] + 0.5870 * image2[:,:,1] + 0.1140 * image2[:,:,2])

distances = [1]
angles = [0]
properties = ['energy', 'contrast', 'dissimilarity', 'homogeneity', 'correlation', 'ASM']

glcm1 = greycomatrix(image1_gray, distances=distances, angles=angles, levels=256, symmetric=True, normed=True)
features1 = [greycoprops(glcm1, prop)[0, 0] for prop in properties]

glcm2 = greycomatrix(image2_gray, distances=distances, angles=angles, levels=256, symmetric=True, normed=True)
features2 = [greycoprops(glcm2, prop)[0, 0] for prop in properties]

# Menampilkan Hasil
print('Fitur Tekstur untuk Gambar 1:')
for feature_name, feature_value in zip(properties, features1):
    print(f'{feature_name}: {feature_value}')

print('\nFitur Tekstur untuk Gambar 2:')
for feature_name, feature_value in zip(properties, features2):
    print(f'{feature_name}: {feature_value}')
```
